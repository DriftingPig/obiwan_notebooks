```python
%pylab inline
import matplotlib.pyplot as plt
import astropy.io.fits as fits
import numpy as np

topdir = '/global/cscratch1/sd/huikong/obiwan_Aug/repos_for_docker/obiwan_out/eboss_ccds/'
ccd1=topdir+'survey-ccds-dr3.fits.gz'
ccd2=topdir+'survey-ccds-dr3_utah_sgc.fits.gz'
ccd1=fits.getdata(ccd1)
ccd2=fits.getdata(ccd2)
ccd1 = ccd1[(ccd1['dec']<2.3)&(ccd1['ra']>200)&(ccd1['dec']>-2.3)]
ccd2 = ccd2[ccd2['ra']>250]
plt.plot(ccd1['ra'],ccd1['dec'],'k,')
plt.plot(ccd2['ra'],ccd2['dec'],'r,')
```

```python
legacyV=fits.getdata('/global/cscratch1/sd/huikong/obiwan_Aug/repos_for_docker/obiwan_out/eboss_ccds/ELG_brick_infos.fits')
```
